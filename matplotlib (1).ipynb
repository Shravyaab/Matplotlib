```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

y1 = x + 1
y2 = x + 2
y3 = x + 3
y4 = x + 4
y5 = x + 5
y6 = x + 6
y7 = x + 7
y8 = x + 8
y9 = x + 9
y10 = x + 10


ax.plot(x, y1, color='pink', lw=2, alpha=1.0)
ax.plot(x, y2, color='orange', lw=2)
ax.plot(x, y3, color='green', lw=2)
ax.plot(x, y4, color='blue', lw=2)
ax.plot(x, y5, color='blue', lw=2, ls='--', marker='o')
ax.plot(x, y6, color='blue', lw=2, ls='--')
ax.plot(x, y7, color='yellow', lw=2, ls='--', marker='o')
ax.plot(x, y8, color='pink', lw=2, ls='--', marker='o')
ax.plot(x, y9, color='orange', lw=2, ls='--', marker='o')
ax.plot(x, y10, color='pink', lw=2, ls='--', marker='s',
        markersize=10, markerfacecolor='yellow',
        markeredgecolor='green', markeredgewidth=3)

plt.show()
```
